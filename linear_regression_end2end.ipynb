```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### 1. Load Boston House Pricing Dataset

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
print(boston.DESCR)
```

```python
dataset = pd.DataFrame(boston.data, columns=boston.feature_names)
dataset.head()
```

```python
dataset['Price'] = boston.target
dataset.head()
```

```python
dataset.info()
```

```python
# Summarizing the Stats
dataset.describe()
```

```python
# Check for missing values
dataset.isnull().sum()
```

```python
# Exploratory Data Analysis
# Correlation
dataset.corr()
```

```python
import seaborn as sns
sns.pairplot(dataset)
```

```python
plt.scatter(dataset['CRIM'], dataset['Price'])
plt.xlabel("Crime Rate")
plt.ylabel("Price")
plt.show()
```

```python
plt.scatter(dataset['RM'], dataset['Price'])
plt.xlabel("No. of Rooms")
plt.ylabel("Price")
plt.show()
```

```python
sns.regplot(x='RM', y='Price', data=dataset)
```

```python
sns.regplot(x='CHAS', y='Price', data=dataset)
```

```python
sns.regplot(x='PTRATIO', y='Price', data=dataset)
```

```python
# Independent and Dependent Features
X = dataset.iloc[:,:-1]
y = dataset.iloc[:, -1]
```

```python
X.head()
```

```python
y.head()
```

```python
# Train-Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.head()
```

```python
# Standard Scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

### Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
regression.fit(X_train, y_train)
```

```python
# Print coefficients and the intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
regression.get_params()
```

```python
reg_pred = regression.predict(X_test)
```

```python
plt.scatter(y_test, reg_pred)
plt.xlabel("Y Test")
plt.ylabel("Y Prediction")
```

```python
# Residuals
residuals = y_test - reg_pred
```

```python
sns.displot(residuals, kind="kde")
```

```python
# Scatter plot with respect to the prediction and residuals
plt.scatter(reg_pred, residuals)
```

### Performance Metrics

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

print("MAE = ", mean_absolute_error(y_test, reg_pred))
print("MSE = ", mean_squared_error(y_test, reg_pred))
print("RMSE = ", np.sqrt(mean_squared_error(y_test, reg_pred)))
```

```python
# R Square
from sklearn.metrics import r2_score
score = r2_score(y_test, reg_pred)
print(score)
```

```python
# Adjusted R square (R square adjusted to the number of predictors/features)
1 - (1 - score) * (len(y_test) - 1) / (len(y_test) - X_test.shape[1] - 1)
```
